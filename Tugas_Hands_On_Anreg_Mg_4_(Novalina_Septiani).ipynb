**TUGAS 1**

```python
import numpy as np
from scipy.stats import pearsonr
import pandas as pd
import os
from google.colab import drive
import matplotlib.pyplot as plt
```

**DATA**

```python
# Load dataset (pastikan sudah upload di Colab)
df = pd.read_csv("StudentsPerformance.csv")
```

**KOEFISIEN KORELASI PEARSON**

```python
# Ambil variabel
variabel_x = df["reading score"]
variabel_y = df["writing score"]

print("Data Variabel X (Reading Score):")
print(variabel_x.head())

print("\nData Variabel Y (Writing Score):")
print(variabel_y.head())

# Hitung korelasi Pearson
korelasi_pearson, p_value = pearsonr(variabel_x, variabel_y)

print(f"\nKoefisien Korelasi Pearson: {korelasi_pearson:.3f}")
print(f"p-value: {p_value:.5f}")

# Interpretasi hasil
if korelasi_pearson > 0.8:
    interpretasi = "Korelasi sangat kuat positif"
elif korelasi_pearson > 0.6:
    interpretasi = "Korelasi kuat positif"
elif korelasi_pearson > 0.4:
    interpretasi = "Korelasi sedang positif"
elif korelasi_pearson > 0.2:
    interpretasi = "Korelasi lemah positif"
elif korelasi_pearson > 0:
    interpretasi = "Korelasi sangat lemah positif"
elif korelasi_pearson < -0.8:
    interpretasi = "Korelasi sangat kuat negatif"
elif korelasi_pearson < -0.6:
    interpretasi = "Korelasi kuat negatif"
elif korelasi_pearson < -0.4:
    interpretasi = "Korelasi sedang negatif"
elif korelasi_pearson < -0.2:
    interpretasi = "Korelasi lemah negatif"
elif korelasi_pearson < 0:
    interpretasi = "Korelasi sangat lemah negatif"
else:
    interpretasi = "Tidak ada korelasi linier"

print("\nInterpretasi Kekuatan Korelasi:", interpretasi)
```

**SCATTERPLOT UNTUK KORELASI TERKUAT**

```python
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn
# Buat scatterplot
plt.figure()
sns.regplot(x=variabel_x, y=variabel_y)
# Tambahkan label dan judul
plt.title("Scatterplot Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
# Tampilkan plot
plt.show()
```

**UJI HIPOTESIS KORELASI TERKUAT**

```python
# Hipotesis
# H0 : Tidak ada hubungan linier antara Reading Score dan Writing Score ( = 0)
# H1 : Ada hubungan linier antara Reading Score dan Writing Score (  0)

alpha = 0.05

print("\n=== UJI HIPOTESIS KORELASI PEARSON ===")
print(f"Nilai r: {korelasi_pearson:.3f}")
print(f"P-value: {p_value:.5f}")
print(f"Alpha: {alpha}")

# Keputusan
if p_value < alpha:
    keputusan = "Tolak H0"
    interpretasi = "Terdapat hubungan linier yang signifikan antara Reading Score dan Writing Score."
else:
    keputusan = "Gagal Menolak H0"
    interpretasi = "Tidak terdapat hubungan linier yang signifikan antara Reading Score dan Writing Score."

print("\nKeputusan:", keputusan)
print("Interpretasi:", interpretasi)

# Interpretasi kekuatan korelasi terkuat
if abs(korelasi_pearson) > 0.8:
    kekuatan = "Sangat Kuat"
elif abs(korelasi_pearson) > 0.6:
    kekuatan = "Kuat"
elif abs(korelasi_pearson) > 0.4:
    kekuatan = "Sedang"
elif abs(korelasi_pearson) > 0.2:
    kekuatan = "Lemah"
else:
    kekuatan = "Sangat Lemah"

# Arah hubungan
if korelasi_pearson > 0:
    arah = "Positif (Searah)"
elif korelasi_pearson < 0:
    arah = "Negatif (Berlawanan arah)"
else:
    arah = "Tidak ada arah hubungan"

print("Kekuatan Hubungan:", kekuatan)
print("Arah Hubungan:", arah)
```

**TUGAS 2**

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np

# Load dataset
data = pd.read_csv("StudentsPerformance.csv")

# Pilih variabel (contoh: reading vs writing)
x = data["reading score"]
y = data["writing score"]
```

**KOEFISIEN KORELASI PEARSON**

```python
# Hitung koefisien korelasi Pearson dan p-value
r, p_value = pearsonr(x, y)

# Tetapkan tingkat signifikansi (alpha)
alpha = 0.05
print(f"Koefisien Korelasi pearson (r): {r:.3f}")
print(f"P-value: {p_value:.5f}")
print(f"Tingkat Signifikansi (Alpha): {alpha}")

# Uji hipotesis
if p_value < alpha:
    print("\nKeputusan: Tolak Hipotesis Nol H0")
    print("Terdapat hubungan linier yang signifikan antara Variabel X dan Variabel Y.")
else:
    print("\nKeputusan: Gagal Menolak Hipotesis Nol H0")
    print("Tidak ada bukti statistik yang cukup untuk  menyimpulkan adanya hubungan linier yang signifikan antara Variabel X dan Variabel Y .")
```

**SCATTERPLOT UNTUK KORELASI TERKUAT**

```python
# Buat scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(variabel_x, variabel_y)

# Tambahkan label dan judul
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')

# Hitung garis regresi linear
m, b = np.polyfit(variabel_x, variabel_y, 1)

# Tambahkan garis regresi
plt.plot(variabel_x, m*variabel_x + b, color='red', linewidth=2)

plt.title('Scatterplot Korelasi antara Reading Score dan Writing Score')
plt.grid(True)

# Tampilkan plot
plt.show()
```
